```python
#Step 1. The first step into creating a waffle chart is determing the proportion of each category with respect to the total.

total_val = np.sum(wafData['Total'])
print(total_val)

##find proportion of each category
prop_cat = [float(x) / total_val for x in wafData['Total']]
print(prop_cat)

for i, prop_val in enumerate(prop_cat):
    print(wafData.index.values[i] + ': ' + str(prop_val))
```

```python
#Step 2. The second step is defining the overall size of the waffle chart.
width = 40
height = 10

total_tiles = width * height
print('total tiles ', ' ' + str(total_tiles))
```

```python
#Step 3. The third step is using the proportion of each category to determe it respective number of tiles
numberOfTilesForCat = [ round(float(x) * total_tiles) for x in prop_cat]
numberOfTilesForCat 

for i, cat in enumerate(numberOfTilesForCat):
    print(df.index.values[i] + ': ' + str(cat))
```

```python
#Step 4. The fourth step is creating a matrix that resembles the waffle chart and populating it.
Chart = np.zeros((height, width))

cat_index=0
index_value=0

##filling value to chart
for col in range(width):
    for row in range(height):
        
        index_value += 1
        
        if index_value > sum(numberOfTilesForCat[0:cat_index]):
            cat_index += 1
            
        Chart[row, col] = cat_index
        
print('Waffle chart Populated')  
        
```
