```python
import pandas as pd
import warnings 

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('..\ChurnData.csv')
df.head()
```

```python
df.describe().T
```

```python
df.dtypes
```

```python
#select features for modeling
chrun_df = df[df.columns.tolist()[:7]]
chrun_df.loc[:, 'churn'] = df.loc[:,'churn'].astype('int')
chrun_df.head()
```

```python
chrun_df.describe()
```

```python
x = chrun_df.drop(columns='churn')
x.head()
```

```python
chrun_df.churn.value_counts()
```

```python
y = chrun_df['churn']
y.head()
```

```python
from sklearn.preprocessing import StandardScaler

x = StandardScaler().fit(x).transform(x)
x[0:5]
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=4)
print(f'Train set: {xtrain.shape}  {ytrain.shape}')
print(f'Train set: {xtest.shape}  {ytest.shape}')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, f1_score, precision_score, recall_score
```

```python
lr = LogisticRegression(C=0.01, solver='liblinear')
lr.fit(xtrain, ytrain)
```

```python
yhat = lr.predict(xtest)
yhat[0:5]
```

```python
yhat_prob = lr.predict_proba(xtest)
yhat_prob[0:5]
```

```python
print(f' F1 Score for the model  with Testing data: {f1_score(ytest, yhat)}')
print(f' F1 Score for the model with Training data: {f1_score(ytrain, lr.predict(xtrain))}')
```

```python
print(f'Precision Score for model with Testing Data : {precision_score(ytest, yhat)}')
print(f'Precision Score for model with Training Data : {precision_score(ytrain, lr.predict(xtrain))}')
```

```python
print(f'recall Score for Testing data: {recall_score(ytest, yhat)}')
print(f'recall Score for Training data: {recall_score(ytrain, lr.predict(xtrain))}')
```

```python
# Jaccard Index
from sklearn.metrics import jaccard_similarity_score
jaccard_similarity_score(ytest, yhat)
```

```python
from sklearn.metrics import jaccard_score
jaccard_score(ytest, yhat)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_context(context='poster')
sns.heatmap(confusion_matrix(ytest, yhat), annot=True, linewidths=0.2)
plt.ylabel('True Values')
plt.xlabel('predicted Value')
plt.title('Confusion Matrix')
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, yhat))
```

```python
from sklearn.metrics import log_loss
log_loss(ytest, yhat_prob)
```
